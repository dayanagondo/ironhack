```python
# Organizing the countries in the data

import re

for row in df_shark['country']:
    if isinstance(row, str):
        new_row = re.sub('\/.+|\(.+\)|\.|\?', '', row)
        new_row = re.sub('\&', 'and', new_row.strip().lower())
        
        if new_row == 'usa':
            new_row = new_row.replace(new_row, 'united states')
        elif new_row == 'bahamas':
            new_row = new_row.replace(new_row, 'the bahamas')
        elif new_row == 'england' or new_row == 'british isles':
            new_row = new_row.replace(new_row, 'united kingdom')
        elif new_row == 'reunion':
            new_row = new_row.replace(new_row, 'réunion')
        elif new_row == 'okinawa':
            new_row = new_row.replace(new_row, 'japan')
        elif new_row == 'azores':
            new_row = new_row.replace(new_row, 'portugal')
        elif new_row == 'red sea':
            new_row = new_row.replace(new_row, 'egypt')
        elif new_row == 'okinawa':
            new_row = new_row.replace(new_row, 'japan')
        elif new_row == 'columbia':
            new_row = new_row.replace(new_row, 'colombia')
        elif new_row == 'new britain' or new_row == 'new guinea' or new_row == 'british new guinea' or new_row == 'admiralty islands':
            new_row = new_row.replace(new_row, 'papua new guinea')
        
        df_shark['country'].replace(row,new_row, inplace=True)
    else:
        df_shark['country'].replace(row,np.nan, inplace=True)
```

```python
for row in df_shark_usa_aus_sa['activity']:
    list_of_actions = ['surfing','swimming','fishing','spearfishing','wading','bathing',
                       'standing','diving','snorkeling','scuba diving','body boarding',
                      'floating','activity with shark','standing','kayaking','walking',
                      'walking','playing','jumping','filming',]
    
    if isinstance(row, str):
        new_activity = re.sub('.*,.*|\-|\/.+|\(.+\)|\.|\?', '', row)
        new_activity = re.sub('\&', 'and', new_activity.strip().lower())

        if re.search('.*?surfi?n?g?.*', new_activity) or re.search('surf', new_activity):
            new_activity = 'surfing'
        elif re.search('.*?swimm?i?n?g?.*', new_activity):
            new_activity = 'swimming'
        elif re.search('.*?fishi?n?g?.*',new_activity):
            new_activity = 'fishing'
        elif re.search('.*?spearfishi?n?g?.*',new_activity):
            new_activity = 'spearfishing'
        elif re.search('.*?wadi?n?g?.*',new_activity):
            new_activity = 'wading'
        elif re.search('.*?bathi?n?g?.*',new_activity):
            new_activity = 'bathing'
        elif re.search('.*?standi?n?g?.*',new_activity):
            new_activity = 'standing'
        elif re.search('.*?divi?n?g?.*',new_activity):
            new_activity = 'diving'
        elif re.search('.*?snorkeli?n?g?.*',new_activity):
            new_activity = 'snorkeling'
        elif re.search('.*?scuba diving.*',new_activity): 
            new_activity = 'scuba diving'
        elif re.search('.*?body boardi?n?g?.*',new_activity) or re.search('.*?body\s?boardi?n?g?.*',new_activity) : 
            new_activity = 'body boarding'
        elif re.search('.*?floati?n?g?.*',new_activity): 
            new_activity = 'floating'
        elif re.search('.*?shark.*',new_activity): 
            new_activity = 'activity with shark'
        elif re.search('.*standi?n?g.*',new_activity): 
            new_activity = 'standing'
        elif re.search('.*kayaki?n?g?.*',new_activity): 
            new_activity = 'kayaking'
        elif re.search('.*walki?n?g?.*',new_activity):
            new_activity = 'walking'
        elif re.search('.*playi?n?g?.*',new_activity):
            new_activity = 'playing'
        elif re.search('.*jumpi?n?g?.*',new_activity):
            new_activity = 'jumping'
        elif re.search('.*filmi?n?g?.*',new_activity):
            new_activity = 'filming'
        elif re.search('.*row?n?g?.*',new_activity):
            new_activity = 'rowing'
        elif new_activity not in list_of_actions:
            new_activity = 'something else'
        
            
        df_shark_usa_aus_sa['activity'].replace(row,new_activity, inplace=True)
    else:
        df_shark_usa_aus_sa['activity'].replace(row,'other', inplace=True)
```

```python
#Alterando os valores errados na coluna date
def change_dates(date):
    import unidecode
    date = unidecode.unidecode(date).lower()
    
    if re.search('.*\-*jan\-*', date):
        return re.sub('.*\-*jan\-*', 'jan', date, flags=re.IGNORECASE)
    if re.search('.*\-*feb\-*', date):
        return re.sub('.*\-*feb\-*', 'feb', date, flags=re.IGNORECASE)
    if re.search('.*\-*mar\-*', date):
        return re.sub('.*\-*mar\-*', 'mar', date, flags=re.IGNORECASE)
    if re.search('.*\-*apr\-*', date):
        return re.sub('.*\-*apr\-*', 'apr', date, flags=re.IGNORECASE)
    if re.search('.*\-*may\-*', date):
        return re.sub('.*\-*may\-*', 'may', date, flags=re.IGNORECASE)
    if re.search('.*\-*jun\-*', date):
        return re.sub('.*\-*jun\-*', 'jun', date, flags=re.IGNORECASE)
    if re.search('.*\-*jul\-*', date):
        return re.sub('.*\-*jul\-*', 'jul', date, flags=re.IGNORECASE)
    if re.search('.*\-*aug\-*', date):
        return re.sub('.*\-*aug\-*', 'aug', date, flags=re.IGNORECASE)
    if re.search('.*\-*sep\-*', date):
        return re.sub('.*\-*sep\-*', 'sep', date, flags=re.IGNORECASE)
    if re.search('.*\-*oct\-*', date):
        return re.sub('.*\-*oct\-*', 'oct', date, flags=re.IGNORECASE)
    if re.search('.*\-*nov\-*', date):
        return re.sub('.*\-*nov\-*', 'nov', date, flags=re.IGNORECASE)
    if re.search('.*\-*dec\-*', date):
        return re.sub('.*\-*dec\-*', 'dec', date, flags=re.IGNORECASE)
    if re.search('.*summ?er.*', date):
        return re.sub('.*summ?er.*', 'summer', date, flags=re.IGNORECASE)
    if re.search('.*winter.*', date):
        return re.sub('.*winter.*', 'winter', date, flags=re.IGNORECASE)
    if re.search('.*spring.*', date):
        return re.sub('.*spring.*', 'spring', date, flags=re.IGNORECASE)
    if re.search('.autumn.*', date):
        return re.sub('.autumn.*', 'autumn', date, flags=re.IGNORECASE)
    else:
        return date
```

```python
df_analysis['date'] = df_analysis['date'].apply(change_dates)
```

```python
#Alterando os valores errados na coluna area
def change_names(name):
    import unidecode
    string = unidecode.unidecode(name).lower()
    
    if re.search('.*[Qq]ueensland.*', string):
        return re.sub('.*[Qq]ueensland.*', 'queensland', string, flags=re.IGNORECASE)
    if re.search('.*[Tt]orres Strait.*', string):
        return re.sub('.*[Tt]orres Strait.*', 'torres strait', string, flags=re.IGNORECASE)
    if re.search('.*[Vv]ictoria.*', string):
        return re.sub('.*[Vv]ictoria.*', 'victoria', string, flags=re.IGNORECASE)
    if re.search('.*[Ee]astern [Cc]ape  [Pp]rovince.*', string):
        return re.sub('.*[Ee]astern [Cc]ape  [Pp]rovince.*', 'eastern cape province', string, flags=re.IGNORECASE)
    if re.search('.*[Nn]ew [Jj]ersey.*', string):
        return re.sub('.*[Nn]ew [Jj]ersey.*', 'new jersey', string, flags=re.IGNORECASE)
    if re.search('.*[Nn]orth [Cc]arolina.*', string):
        return re.sub('.*[Nn]orth [Cc]arolina.*', 'north carolina', string, flags=re.IGNORECASE)
    if re.search('.*[Ff]lorida.*', string):
        return re.sub('.*[Ff]lorida.*', 'florida', string, flags=re.IGNORECASE)
    if re.search('.*[Nn]ew [Yy]ork.*', string):
        return re.sub('.*[Nn]ew [Yy]ork.*', 'new york', string, flags=re.IGNORECASE)
    if re.search('.*[Ss]outh [Cc]arolina.*', string):
        return re.sub('.*[Ss]outh [Cc]arolina.*', 'south carolina', string, flags=re.IGNORECASE)
    if re.search('Westerm Australia', string):
        return re.sub('Westerm Australia', 'western australia', string, flags=re.IGNORECASE)
    if re.search('CUBA', string):
        return re.sub('CUBA', 'cuba', string, flags=re.IGNORECASE)
    else:
        return string
```
